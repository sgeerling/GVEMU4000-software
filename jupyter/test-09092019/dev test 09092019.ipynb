```python
from sqlalchemy import create_engine                                                                                                                                    
                                                                                                                                                       
                                                                                                                                                                        
class Database:                                                                                                                                                         
    def __init__(self,                                                                                                                                                  
                 db_name="test_gopimn",                                                                                                                       
                 username="sa",                                                                                                                
                 pw="eTrans2019",                                                                                                                            
                 host="localhost",                                                                                                                        
                 port=1433):                                                                                                                  
                                                                                                                                                                        
        self.engine = create_engine('mssql+pymssql://{}:{}@{}:{}/{}'.format(username, pw, host, port, db_name))                                                         
                                                                                                                                                                        
                                                                                                                                     
db = Database()                                                                                                                                                     
                      
```

```python
from sqlalchemy import MetaData
from sqlalchemy import Table
from sqlalchemy import Column
from sqlalchemy import Integer
from sqlalchemy import String
from sqlalchemy import DateTime
from sqlalchemy import Float

truck_geo=[]
raw_all = []
t_stamp=[]
try:
    #sql_query = "select top 100 * from test_queue_raw where raw_data like '%+RESP:GTFRI,270601,865067022303617%' and Id >= 73515"
    sql_query = "select * from test_queue_raw where raw_data like '+RESP:GTFRI,270601,865067022303617%'"# and Id >= 73515"
    cn = db.engine.connect()
    result = cn.execute(sql_query).fetchall()
    for row in result:  
        raw = row[1]
        frame = raw.split(",")
        raw_all += raw
        try:
            if frame[12]!='nan'  and frame[11]!='nan':
                curr_geo = [float(frame[12]) , float(frame[11])]
                truck_geo += [curr_geo]
                t_stamp.append(frame[13])
        except:
            print("bad ")
    cn.close()
except Exception as e:
    cn.close()
    raise e
```

```python
geor = []
t_stamp[0]
gtfri_info= []
for i in range(0,len(truck_geo)):
    if t_stamp[i] >= '20190909212000' and t_stamp[i] <= '20190910120000' :
        gtfri_info.append({'datetime':t_stamp[i], 'location':truck_geo[i]})
        geor += [truck_geo[i]]
geor
```

```python
info_box_template = """
<dl>
<dt>time stamp</dt><dd>{datetime}</dd>
</dl>
"""
datetime_info = [info_box_template.format(**datetime) for datetime in gtfri_info]
datetime_info
```

```python
fig = gmaps.figure()
marker_layer = gmaps.marker_layer(geor, info_box_content=datetime_info)
fig.add_layer(marker_layer)
fig
```
